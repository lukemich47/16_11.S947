```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
%matplotlib inline
```

```python
jsonlist = []
for x in range (250,535):
    jsonlist.append("jsons-foursquare2015-10-09/" + str(x) + ".json")
jsonlist
```

```python
df_concat = pd.read_json(jsonlist[0])
df_concat
    
```

```python

for x in range(1,285):
    df = pd.read_json(jsonlist[x])
    df_concat = pd.concat([df_concat,df], axis=1)
df_concat
```

```python
df_con_trans = df_concat.transpose()
df_con_trans.head()
```

```python
df_con_trans[['checkins', 'here_now', 'users_count', 'lat', 'lon']] = df_con_trans[['checkins', 'here_now', 'users_count', 'lat', 'lon']].astype('float')
```

```python
category_checkins_users = df_con_trans[['category','checkins','users_count']].groupby('category').mean()
category_checkins_users
```

```python
category_checkins_users['combined'] = category_checkins_users['checkins'] + category_checkins_users['users_count']
category_checkins_users = category_checkins_users.sort_values('combined', ascending=False)
category_checkins_users
```

```python


fig, ax = plt.subplots(figsize=(15,10))
ax = category_checkins_users[['checkins', 'users_count']].plot(kind='bar', title='Checkins & User Counts by Category', ax=ax, stacked=True)
ax.set_ylabel('Mean Count')
ax.set_xlabel('Place Category')
fig.tight_layout()
plt.show()
```

```python
place_locations = df_con_trans[['place_id','checkins','users_count', 'lat','lon']].groupby('place_id').mean()
place_locations
```

```python


fig, ax = plt.subplots(figsize=(15,10))
ax = place_locations.plot(kind='scatter', x='lon', y='lat', 
                          title='Place Locations, Checkins (size), and User Counts (color)', 
                          s=place_locations['checkins']*.1, c=place_locations['users_count'], colormap = 'Blues', ax=ax)
ax.set_ylabel('Latitude')
ax.set_xlabel('Longitude')
fig.tight_layout()
plt.show()
```
